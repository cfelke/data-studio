```python
all_states=[]
for state in abbreviations:
    url = 'https://locator.pregnancydecisionline.org/WS/api/GetPCResults?ClientID=7961897b-815b-4161-b0c6-c9350dfde095&ZipCode=&StateProvince='+ state +'&Style=style1'
    try:
        temp = requests.get(url).json()
        all_states.append(temp)
    except:
        pass
```

```python
all_states_dict = []
for s in [state['Results'] for state in all_states]:
    all_states_dict = all_states_dict + s
```

```python
df = pd.DataFrame(all_states_dict)
```

```python
from cycler import cycler
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def get_color(CPC):
    #print(CPC)
    if CPC > 10:
        return '#96c93b'
    else:
        return '#4f9c8b'
    
bar_colors = [get_color(city) for city in CPC_citites.sort_values()]
ax = CPC_citites.sort_values().plot(kind='barh', y='City', figsize=(8,6), color=bar_colors, legend=False)

for i, bar in enumerate(ax.patches):
    # print(i)
    value = int(round(bar.get_width()))
    x_pos = bar.get_width()-0.25
    y_pos = i
    # ax.text(x_pos, y_pos, value, verticalalignment='center')
    ax.text(x_pos, y_pos, value, va='center', ha='right', color='white', fontweight='bold')
    
ax.set_title('Most CPC are located in Columbus, Ohio', pad=30)
ax.title.set_fontsize(20)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)

#This removes the ticks on the left, grid_alpha changes the opacity
ax.tick_params(which='major', left=False, bottom=False, grid_alpha=0.3)

#ax.yaxis.label.set_visible(False)
plt.rcdefaults()
plt.rcParams.update({
  'pdf.fonttype': 42,
  'ps.fonttype': 42,
  'font.size': 10.0
})

plt.show()
plt.savefig('CPC_draft.pdf')
```
