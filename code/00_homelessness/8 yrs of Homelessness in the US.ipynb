```python
from functools import reduce
```

```python
dfs = [df_2010,df_2011, df_2012, df_2013, df_2014, df_2015, df_2016, df_2017]

df_homeless_8yrs = reduce(lambda left,right: pd.merge(left,right,on='State'), dfs)
```

```python
df_total_8yrs['Percentage Homelessness, 2017'] = df_total_8yrs['Total Homeless, 2017']/df_total_8yrs['Population Estimate 2017'] * 100
df_total_8yrs.head()
```

```python
ax = df_total_8yrs.sort_values(by=['Percentage Homelessness, 2017']).plot(kind='barh', x='US_State', y='Percentage Homelessness, 2017', figsize=(10,30))
ax.set_title('In 2017 most homeless people lived in the District of Columbia', pad=30)
ax.title.set_fontsize(20)
```

```python
#Show the overall development of homelessness in the US since 2010

df_total_8yrs['Total Homeless, 2010'].sum()
```

```python
years = {}
n = 2010
while n < 2018:
    column = 'Total Homeless, '+str(n)
    years[n]=df_total_8yrs[column].sum()
    n+=1
```

```python
index_ = range(2010,2018)
df_years = pd.DataFrame.from_dict(years, orient='index')
```

```python
import matplotlib.font_manager
from IPython.core.display import HTML

def make_html(fontname):
    return "<p>{font}: <span style='font-family:{font}; font-size: 24px;'>{font}</p>".format(font=fontname)

code = "\n".join([make_html(font) for font in sorted(set([f.name for f in matplotlib.font_manager.fontManager.ttflist]))])

HTML("<div style='column-count: 2;'>{}</div>".format(code))
```

```python
plt.rcdefaults()

ax = df_years.plot(kind='line', color='c', figsize=(8, 5), legend=False)
ax.set_title('Homelessness in the US was slowly (very slowly) decreasing. Not anymore.', pad=50)
ax.title.set_fontsize(25)
ax.title.set_fontname("Myanmar Text")
ax.set_ylabel('Number of Homeless')

ax.set_xticks([2010, 2012, 2014, 2016,])
ax.set_yticks([0, 100000,200000, 300000, 300000, 400000, 500000, 600000, 700000])

ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['right'].set_visible(False)

# #This removes the ticks on the left, grid_alpha changes the opacity
ax.tick_params(which='major', left=False, bottom=False, grid_alpha=0.5)

ax.grid(True, axis='y')
ax.set_axisbelow(True)

plt.savefig("simple_line_homelessness.pdf")
```
