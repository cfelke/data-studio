```python
df_total = df[:55]
df_total
```

```python
df_pop = pd.read_excel('US_pop_2017-01.xlsx')
df_pop.head(15)
```

```python
df_pop.dtypes
```

```python
pop_clean = df_pop[8:61]
```

```python
pop_clean.drop(columns=['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9'])
```

```python
pop_clean.rename(columns={'table with row headers in column A and column headers in rows 3 through 4. (leading dots indicate sub-parts)':'State_Name', 'Unnamed: 10':'Total Population'}, inplace=True)
pop_clean.head()
```

```python
pop_final = pop_clean.drop(columns=['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9'])
pop_final['State'] = pop_final['State_Name'].str.extract(r"[.]?(.*)")
pop_final.shape
```

```python
df_homeless = pop_final.merge(df_total,on='State')
df_homeless.shape
```

```python
df_homeless.drop(columns={'State_Name'}, inplace=True)
df_homeless
```

```python
df_homeless['Percentage Homelessness'] = df_homeless['Total Homeless, 2017']/df_homeless['Total Population'] * 100
df_homeless.head()
```

```python
ax = df_homeless.sort_values(by=['Percentage Homelessness']).plot(kind='barh', x='State', y='Percentage Homelessness', figsize=(10,30))
ax.set_title('Most homeless people live in the District of Columbia', pad=30)
ax.title.set_fontsize(20)
```

```python
df_homeless_five = df_homeless.sort_values(by=['Percentage Homelessness'], ascending=False).head(5)
df_homeless_five
```

```python
df_homeless_five.dtypes
```

```python
import matplotlib.font_manager
from IPython.core.display import HTML

def make_html(fontname):
    return "<p>{font}: <span style='font-family:{font}; font-size: 24px;'>{font}</p>".format(font=fontname)

code = "\n".join([make_html(font) for font in sorted(set([f.name for f in matplotlib.font_manager.fontManager.ttflist]))])

HTML("<div style='column-count: 2;'>{}</div>".format(code))
```

```python
def get_color(homeless):
    #print(homeless)
    if homeless > 1:
        return 'purple'
    else:
        return 'lightgrey'
    
bar_colors = [get_color(percentage) for percentage in df_homeless_five['Percentage Homelessness'].sort_values()]

ax = df_homeless_five.sort_values(by=['Percentage Homelessness']).plot(kind='barh', color=bar_colors, x='State', y='Percentage Homelessness', figsize=(10,4), legend = False)
ax.set_title('Most homeless people live in the District of Columbia', pad=30)
ax.title.set_fontsize(30)
ax.title.set_fontname("David")

ax.set_xticks([0, 0.5, 1])
ax.set_xlabel('% of Total Population')

ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)

#This removes the ticks on the left, grid_alpha changes the opacity
ax.tick_params(which='major', left=False, bottom=False, grid_alpha=0.3)

#ax.grid(True)
#Put the grid behind
#ax.set_axisbelow(True)

#Just the x-grid
ax.grid(True, axis='x')
ax.set_axisbelow(True)
ax.set_xlim(0, 1.5)

#ax.yaxis.label.set_visible(False)
```
