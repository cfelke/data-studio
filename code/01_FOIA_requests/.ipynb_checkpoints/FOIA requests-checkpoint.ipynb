```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
requests = pd.read_excel('FOIA_EPA_requests_2008_2017.xlsx', sheet_name='requests')
requests.head()
```

```python
exemptions = pd.read_excel('FOIA_EPA_requests_2008_2017.xlsx', sheet_name='exemptions')
exemptions.head()
```

```python
re_exe = requests.merge(exemptions, on= 'Component')
re_exe
```

```python
b3_exemptions = pd.read_excel('FOIA_EPA_requests_2008_2017.xlsx', sheet_name='b3_exemptions')
b3_exemptions.head()
```

```python
b3_exemptions.dtypes
```

```python
re_exe_b3 = re_exe.merge(b3_exemptions, on='Component')
re_exe_b3
```

```python
backlog = pd.read_excel('FOIA_EPA_requests_2008_2017.xlsx', sheet_name='backlog')
backlog.head()
```

```python
re_exe_b3_back = re_exe_b3.merge(backlog, on='Component')
re_exe_b3_back
```

```python
granted_denied = pd.read_excel('FOIA_EPA_requests_2008_2017.xlsx', sheet_name='granted_denied')
granted_denied.head()
```

```python
foia = re_exe_b3_back.merge(granted_denied, on='Component')
foia
```
